```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importando dados do dataset da UNESCO sobre 
dados = pd.read_csv('/content/SCN_DS_15042023073626036.csv')
dados
```

# Nova seção

```python
#Filtrando os dados para o ano de 2021
filtered_values = dados[dados['TIME'] == 2021]
filtered_values
```

```python
#Filtrando os dados de forma descrescente
filtered_values = filtered_values.sort_values(by="Value",ascending=False)
filtered_values
```

```python
#Filtrando os 10 países que mais investiram em pesquisa
filtered_values=filtered_values[:5]
filtered_values
```

```python
Country = filtered_values['Country']
```

```python
Value = filtered_values['Value']
```

```python
Cores = ['black', 'dimgray', 'gray', 'grey', 'lightgrey']
```

```python
plt.title('Países que mais investiram em pesquisa')
plt.rcParams["figure.figsize"] = (25,8)
plt.bar(Country, Value, align='center',  width=0.5, color=Cores)
plt.xlabel ('Países')
plt.ylabel('Valor')
plt.show()
```

Conclusão: A China é o país que mais investe em pesquisa, ocupando a primeira e segunda posição através de suas regiões administrativas
